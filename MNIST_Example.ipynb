# 1. The MNIST Dataset

MNIST is one of the most famous datasets in machine learning.

- 60,000 training images
- 10,000 test images
- 10 classes: digits 0–9
- Each image has size **28 × 28** pixels

Thus each image contains

$$
28 \times 28 = 784
$$

pixel values.

# 3. Image as a Matrix

Mathematically, the image is a matrix:

$$
X =
\begin{bmatrix}
x_{1,1} & x_{1,2} & \cdots & x_{1,28}\\
x_{2,1} & x_{2,2} & \cdots & x_{2,28}\\
\vdots & \vdots & \ddots & \vdots\\
x_{28,1} & x_{28,2} & \cdots & x_{28,28}
\end{bmatrix}
$$

Each pixel is a feature.

Example:

$$
x_{3,11}=5
$$

means:

- row = 3
- column = 11
- pixel intensity = 5

---

# 4. Flatten the Image

Neural networks usually expect a vector rather than a matrix.

Therefore we flatten the image row by row:

$$
X=(x_1,x_2,\ldots,x_{784})
$$

Mapping:

```text
x1   = Row 1, Column 1
x2   = Row 1, Column 2
...
x28  = Row 1, Column 28
x29  = Row 2, Column 1
...
x784 = Row 28, Column 28
```

The beginning of the vector might look like:

```text
X =
[0,0,0,0,...,5,9,9,9,9,9,5,...]
```

# 5. Neural Network Architecture

For illustration we use a small network:

```text
784 → 4 → 3 → 10
```

meaning:

```text
784 input neurons
4 neurons in hidden layer 1
3 neurons in hidden layer 2
10 output neurons
```

Diagram:

```text
784 pixels
      ↓
Hidden Layer 1 (4 neurons)
      ↓
Hidden Layer 2 (3 neurons)
      ↓
Output Layer (10 digits)
```

---

# 6. Hidden Layer 1: Feature Detection

The first hidden layer learns simple visual features.

We define:

| Neuron | Learned Feature |
|--------|----------------|
| A1 | Top curve detector |
| A2 | Diagonal detector |
| A3 | Bottom line detector |
| A4 | Loop detector |

Each neuron computes

$$
z_k=b_k+\sum_{j=1}^{784}w_{kj}x_j
$$

and applies ReLU activation:

$$
A_k=\max(0,z_k)
$$

---

## Neuron A1: Top Curve Detector

Digit 2 has a strong top curve.

Suppose

$$
z_1=7.2
$$

After ReLU:

$$
A_1=7.2
$$

Interpretation:

```text
Top curve detected strongly.
```

---

## Neuron A2: Diagonal Detector

Digit 2 contains a diagonal stroke.

Suppose

$$
z_2=8.5
$$

Thus

$$
A_2=8.5
$$

Interpretation:

```text
Strong diagonal stroke detected.
```

---

## Neuron A3: Bottom Line Detector

Digit 2 has a bottom horizontal line.

Suppose

$$
z_3=6.8
$$

Thus

$$
A_3=6.8
$$

Interpretation:

```text
Bottom line detected.
```

---

## Neuron A4: Loop Detector

Digit 2 usually does not have a closed loop.

Suppose

$$
z_4=-1.5
$$

After ReLU:

$$

$$

Interpretation:

```text
No loop detected.
```

# 8. Hidden Layer 2: Combining Features

The second hidden layer combines simple features into digit patterns.

| Neuron | Meaning |
|--------|---------|
| B1 | Looks like digit 2 |
| B2 | Looks like digit 7 |
| B3 | Looks like digit 8 |

---

## Neuron B1: "Looks Like 2"

A digit 2 requires:

- top curve
- diagonal stroke
- bottom line
- no loop

Suppose

$$
z_{B1}
=
0.5
+0.8A_1
+0.9A_2
+0.7A_3
-0.6A_4
$$

Substituting values:

$$
z_{B1}
=
0.5
+0.8(7.2)
+0.9(8.5)
+0.7(6.8)
-0.6(0)
$$

$$
z_{B1}=18.67
$$

Thus

$$
B_1=18.67
$$

Strong evidence for digit 2.

---

## Neuron B2: "Looks Like 7"

Suppose

$$
z_{B2}
=
0.2
+0.7A_1
+0.6A_2
-0.8A_3
$$

Calculating:

$$
z_{B2}=4.90
$$

Moderate evidence for digit 7.

---

## Neuron B3: "Looks Like 8"

Digit 8 requires loops.

Suppose

$$
z_{B3}
=
-1
+0.3A_1
+0.2A_2
+1.5A_4
$$

Since

$$
A_4=0
$$

we obtain

$$
z_{B3}=2.86
$$

Weak evidence for digit 8.

---

# 9. Output of Hidden Layer 2

The second hidden layer outputs

$$
A^{(2)}=
[18.67,\;4.90,\;2.86]
$$

Interpretation:

```text
Very strong evidence for digit 2.
```

|------|------:|

| 2 | 12.5 |
| 3 | 3.0 |
| 4 | -0.5 |
| 5 | 1.1 |
| 6 | -0.8 |

The highest score is for digit **2**.

# 11. Softmax: Convert Scores to Probabilities

Softmax computes:

$$
P(Y=m|X)
=
\frac{e^{z_m}}
{\sum_{\ell=0}^{9}e^{z_\ell}}
$$

Suppose we obtain:

| Digit | Probability |
|------|------------:|
| 0 | 0.000006 |
| 1 | 0.000001 |
| 2 | 0.999600 |
| 3 | 0.000075 |
| 4 | 0.000002 |
| 5 | 0.000008 |
| 6 | 0.000001 |
| 7 | 0.000250 |
| 8 | 0.000028 |
| 9 | 0.000005 |

---

# 12. Final Prediction

The predicted digit is

$$
\hat{Y}=2
$$

with probability

$$
P(Y=2|X)=0.9996
$$

# 13. Summary

```text
28×28 image
      ↓
784 pixel values
      ↓
Hidden Layer 1:
    top curve
    diagonal
    bottom line
    loop
      ↓
Hidden Layer 2:
    digit patterns
      ↓
Output Layer:
    probabilities for 0–9
      ↓
Prediction: digit 2
```

## Key Insight

Deep learning works because it learns **hierarchically**:

```text
pixels
   ↓
edges and curves
   ↓
digit parts
   ↓
whole digits
```
