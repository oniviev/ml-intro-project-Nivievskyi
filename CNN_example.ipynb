# CNN Walkthrough: A 32×32 RGB Image

## Learning Objectives

After this notebook, you should be able to:

- Understand how a color image is represented as data
- Understand RGB channels
- Compute the output dimensions of convolutional layers
- Understand filters, feature maps, and pooling
- Follow a CNN from image to classification

---

# 1. A Color Image as Data

Consider a color image of size

$$
32 \times 32
$$

pixels.

Unlike MNIST, which is grayscale, a color image contains **three channels**:

- Red (R)
- Green (G)
- Blue (B)

Therefore the image dimensions are

$$
32 \times 32 \times 3
$$

The three channels can be visualized as:

```text
Red channel:    32 × 32
Green channel:  32 × 32
Blue channel:   32 × 32
```

Together:

```text
32 × 32 × 3
```

---

# 2. One Pixel

Each pixel contains three numbers:

$$
(R,G,B)
$$

Example:

```text
Pixel (10,15) = (120,80,45)
```

which means

```text
Red   = 120
Green = 80
Blue  = 45
```

This corresponds to a brownish color.

---

# 3. Total Number of Features

The image contains

$$
32 \times 32 \times 3
=
3072
$$

numbers.

Thus a computer initially sees:

```text
3072 numerical values
```

and not a cat, dog, or car.

---

# 4. CNN Input Tensor

A tensor is just a multidimensional array of numbers

Mathematically, the image is a tensor:

$$
X \in \mathbb{R}^{32 \times 32 \times 3}
$$

The three channels are:

$$
X(:,:,1)=\text{Red}
$$

$$
X(:,:,2)=\text{Green}
$$

$$
X(:,:,3)=\text{Blue}
$$

---

# 5. Convolutional Filter

Suppose the CNN uses a filter of size

```text
3 × 3
```

Since the image has 3 channels, the filter must span all channels:

$$
3 \times 3 \times 3
$$

Thus one filter contains

$$
3 \times 3 \times 3 = 27
$$

weights plus one bias.

---

# 6. Visualizing the Filter

A CNN filter is actually three small matrices stacked together:

## Red Channel Weights

$$
W_R=
\begin{bmatrix}
w^{(R)}_{11} & w^{(R)}_{12} & w^{(R)}_{13}\\
w^{(R)}_{21} & w^{(R)}_{22} & w^{(R)}_{23}\\
w^{(R)}_{31} & w^{(R)}_{32} & w^{(R)}_{33}
\end{bmatrix}
$$

## Green Channel Weights

$$
W_G=
\begin{bmatrix}
w^{(G)}_{11} & w^{(G)}_{12} & w^{(G)}_{13}\\
w^{(G)}_{21} & w^{(G)}_{22} & w^{(G)}_{23}\\
w^{(G)}_{31} & w^{(G)}_{32} & w^{(G)}_{33}
\end{bmatrix}
$$

## Blue Channel Weights

$$
W_B=
\begin{bmatrix}
w^{(B)}_{11} & w^{(B)}_{12} & w^{(B)}_{13}\\
w^{(B)}_{21} & w^{(B)}_{22} & w^{(B)}_{23}\\
w^{(B)}_{31} & w^{(B)}_{32} & w^{(B)}_{33}
\end{bmatrix}
$$

The superscript identifies the channel:

```text
(R) = Red
(G) = Green
(B) = Blue
```

while the subscripts identify the position inside the filter.

---

# 7. Extract a Local Patch

The filter looks at a small region of the image (local patch):
suppose the filter is placed on rows 1–3 and columns 1–3 of the image.

The CNN extracts:

## Red patch

$$
X_R=
\begin{bmatrix}
x^{(R)}_{11} & x^{(R)}_{12} & x^{(R)}_{13}\\
x^{(R)}_{21} & x^{(R)}_{22} & x^{(R)}_{23}\\
x^{(R)}_{31} & x^{(R)}_{32} & x^{(R)}_{33}
\end{bmatrix}
$$

## Green patch

$$
X_G=
\begin{bmatrix}
x^{(G)}_{11} & x^{(G)}_{12} & x^{(G)}_{13}\\
x^{(G)}_{21} & x^{(G)}_{22} & x^{(G)}_{23}\\
x^{(G)}_{31} & x^{(G)}_{32} & x^{(G)}_{33}
\end{bmatrix}
$$

## Blue patch

$$
X_B=
\begin{bmatrix}
x^{(B)}_{11} & x^{(B)}_{12} & x^{(B)}_{13}\\
x^{(B)}_{21} & x^{(B)}_{22} & x^{(B)}_{23}\\
x^{(B)}_{31} & x^{(B)}_{32} & x^{(B)}_{33}
\end{bmatrix}
$$

Thus the image patch contains:

```text
9 red values
9 green values
9 blue values
```

for a total of

$$
27
$$

numbers.

---

# 8. Compute One Convolution

The convolution computes the dot product channel by channel:

$$
z
=
\sum_{i=1}^{3}\sum_{j=1}^{3}
w^{(R)}_{ij}x^{(R)}_{ij}
+
\sum_{i=1}^{3}\sum_{j=1}^{3}
w^{(G)}_{ij}x^{(G)}_{ij}
+
\sum_{i=1}^{3}\sum_{j=1}^{3}
w^{(B)}_{ij}x^{(B)}_{ij}
+b
$$

This produces a single scalar:

$$
z
$$

Interpretation:

> How strongly is this visual pattern present in this local region?

---


# 9. Slide the Filter Across the Image

The filter moves over the image:

```text
slide → compute
slide → compute
slide → compute
...
```

Using:

```text
Filter size = 3×3
Stride = 1
Padding = 0
```

The output size becomes

$$
32-3+1=30
$$

Thus one filter produces

$$
30 \times 30 \times 1
$$

feature map.

---

# 10. Multiple Filters

Suppose we use

```text
16 filters
```

Each filter produces one feature map.

Therefore:

$$
32 \times 32 \times 3
\rightarrow
30 \times 30 \times 16
$$

The CNN may learn filters for:

- vertical edges
- horizontal edges
- colors
- textures
- corners
- curves

---

# 11. ReLU Activation

After convolution we apply activation

$$
A=\max(0,z)
$$

If

$$
z=-2.5
$$

then

$$
A=0
$$

If

$$
z=7.8
$$

then

$$
A=7.8
$$

Interpretation:

```text
Large positive value → pattern detected
Zero → pattern absent
```

---

# 12. Max Pooling

Apply 2×2 max pooling:

Input:

```text
4 7
2 5
```

Output:

```text
7
```

Pooling reduces dimensionality while preserving strong signals.

Applying 2×2 pooling:

$$
30 \times 30 \times 16
\rightarrow
15 \times 15 \times 16
$$

---

# 13. Second Convolutional Layer

Now the CNN no longer sees raw RGB pixels.

It sees

$$
15 \times 15 \times 16
$$

feature maps.

Suppose we apply

```text
32 filters
```

Each filter now has dimensions

$$
3 \times 3 \times 16
$$

Number of weights:

$$
3\times3\times16=144
$$

plus one bias.

Output size:

$$
15-3+1=13
$$

Thus:

$$
15 \times 15 \times 16
\rightarrow
13 \times 13 \times 32
$$

---

# 14. Second Pooling Layer

Apply 2×2 pooling:

$$
13 \times 13 \times 32
\rightarrow
6 \times 6 \times 32
$$

approximately.

---

# 15. What Do Deeper Layers Learn?

Layer 1 learns:

```text
edges
colors
textures
corners
```

Layer 2 learns:

```text
eyes
ears
fur
wheels
windows
```

Layer 3 learns:

```text
cat face
car
dog
person
```

CNNs learn hierarchically.

---

# 16. Flattening

Flatten the tensor:

$$
6 \times 6 \times 32
=
1152
$$

Thus:

$$
[feature_1,\ldots,feature_{1152}]
$$

These are learned features, not raw pixels.

---

# 17. Dense Layer

Feed into a fully connected layer:

```text
1152 → 128 neurons
```

The dense layer combines high-level features.

---

# 18. Output Layer

For CIFAR-10 we have 10 classes:

```text
airplane
automobile
bird
cat
deer
dog
frog
horse
ship
truck
```

Softmax converts scores into probabilities:

| Class | Probability |
|------|------------:|
| Cat | 0.72 |
| Dog | 0.18 |
| Deer | 0.04 |
| Horse | 0.02 |
| Others | 0.04 |

Prediction:

$$
\hat{Y}=\text{Cat}
$$

---

# 19. Full CNN Pipeline

```text
Input image
32 × 32 × 3
        ↓
Convolution (16 filters)
30 × 30 × 16
        ↓
ReLU
30 × 30 × 16
        ↓
Max Pooling
15 × 15 × 16
        ↓
Convolution (32 filters)
13 × 13 × 32
        ↓
ReLU
13 × 13 × 32
        ↓
Max Pooling
6 × 6 × 32
        ↓
Flatten
1152
        ↓
Dense Layer
128
        ↓
Softmax
10 classes
```

---

# 20. Key Insight

Unlike ordinary neural networks:

```text
NN:
3072 numbers → prediction
```

CNNs learn hierarchically:

```text
pixels
   ↓
edges
   ↓
textures
   ↓
object parts
   ↓
objects
```

This hierarchical feature extraction is why CNNs dominate computer vision.
